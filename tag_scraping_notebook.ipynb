Function to grab a category of tags, definded by `class_name`

```python
def grab_tag_dict(soup, class_name):
    """
    This function takes in a `soup` variable of type `bs4.BeautifulSoup`,
    obtained from a `tag_url = 'https://archiveofourown.org/tags/' + 'f{tag_name}'`,
    and grabs all the tags listed under under specified `class_name`
    """
    tags_list = soup.find(class_=class_name)
    tags_dict = []
    if tags_list:
        tags_list = tags_list.find_all("li")
        tags_dict = [
            {tag.find("a").string: tag.find("a").attrs["href"]}
            for tag in tags_list
            if tag.find("a")
        ]
    return tags_dict
```

```python
import requests
from bs4 import BeautifulSoup
from lxml import html
```

```python
session_requests = requests.session()
result = session_requests.get(tag_url, headers=dict(referer=tag_url))
print(result.status_code)
soup = BeautifulSoup(result.content, "html.parser")
type(soup)
```
